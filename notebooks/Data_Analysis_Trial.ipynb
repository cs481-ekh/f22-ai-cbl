# Analyis of Univariate Distributions: density analysis

```
sns.displot(dataset,
            x="Max_tib_med_contact_pressure",
            kind="kde"
           );
```

```
sns.displot(dataset,
            x="Max_tib_lat_contact_pressure",
            kind="kde"
           );
```

```
sns.displot(dataset, x="Max_tib_med_contact_pressure", kind="ecdf")
sns.displot(dataset, x="Max_tib_lat_contact_pressure", kind="ecdf")
```
