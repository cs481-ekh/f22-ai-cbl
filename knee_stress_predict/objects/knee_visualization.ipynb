# KNEE GEOMETRY VISUALIZATION
## Setup and Load Data

```python
from pathlib import Path
from knee_stress_predict.config import raw_data_dir
from knee_stress_predict.objects.KneeGeometry import KneeGeometry
```

We will use the KneeGeometry object that creates meshes of 7 elements of the knee based on the folder name located in the data/raw directory (raw_data_dir). This folder is supposed to contain 7 .inp files where each file describes mesh elements of a specified knee part:
- Patella
- Femur
- Tibia
- Patella cartilage
- Femur cartilage (Articular cartilage)
- Tibia medial cartilage
- Tibia lateral cartilage
<img src="../../docs/img/knee_structure.png" alt="drawing" width="500"/>

```python
folder_name = "2022.09.16_Geometries"
path = Path.joinpath(raw_data_dir, folder_name)
knee = KneeGeometry(path)
```

## Mesh visualization of each element
### Femur

```python
print(f"Number of elements: {knee.femur.number_of_cells}")
print(f"Number of points: {knee.femur.number_of_points}")
knee.femur.plot(jupyter_backend="static")
```

### Tibia

```python
print(f"Number of elements: {knee.tibia.number_of_cells}")
print(f"Number of points: {knee.tibia.number_of_points}")
knee.tibia.plot(jupyter_backend="static")
```

### Patella

```python
print(f"Number of elements: {knee.patella.number_of_cells}")
print(f"Number of points: {knee.patella.number_of_points}")
knee.patella.plot(jupyter_backend="static")
```

### Femur cartilage (Articular cartilage)

```python
print(f"Number of elements: {knee.fem_cart.number_of_cells}")
print(f"Number of points: {knee.fem_cart.number_of_points}")
knee.fem_cart.plot(jupyter_backend="static")
```

### Tibia cartilages


```python
print(f"Number of elements Tibia medial cartilage: {knee.tibia_cart_med.number_of_cells}")
print(f"Number of Tibia medial cartilage: {knee.tibia_cart_med.number_of_points}")
print(f"Number of elements Tibia lateral cartilage: {knee.tibia_cart_lat.number_of_cells}")
print(f"Number of points Tibia lateral cartilage: {knee.tibia_cart_lat.number_of_points}")
knee.tibia_cart_med.plot(jupyter_backend="static")
knee.tibia_cart_lat.plot(jupyter_backend="static")
```
